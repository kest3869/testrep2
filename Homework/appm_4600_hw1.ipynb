```python
# libraries
import numpy as np
import matplotlib.pyplot as plt

# the expanded version of f(x)
def y_expanded(x):
  return x**9 - 18*x**8 + 144*x**7 - 672*x**6 \
  + 2016*x**5 - 4032*x**4 + 5376*x**3 - 4608*x**2 \
  + 2304*x - 512

# the compact version of f(x)
def y_compact(x):
  return (x - 2)**9

# initialie x-array
x_arr = np.arange(1.920, 2.080,0.001)

# initialize y_arrays
y_e = np.zeros(len(x_arr))
y_c = np.zeros(len(x_arr))

# calculate values
for i,x in enumerate(x_arr):
  y_e[i] = y_expanded(x)
  y_c[i] = y_compact(x)

# plot the results
plt.plot(x_arr, y_e, label="Expanded")
plt.plot(x_arr, y_c, label="Compact")
plt.title("Expanded vs Compact Plot of (x-2)^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# libraries
import numpy as np
from numpy import cos
from numpy import sin
import matplotlib.pyplot as plt

# the original expression
def y_o(x, d):
  return cos(x+d) - cos(x)

# the re-arranged expression
def y_r(x, d):
  return (cos(x)*cos(d)/sin(x)) \
  + (sin(-1*x)*sin(d)/sin(x)) \
  + cos(-1*x)/sin(-1*x)

# values for x
x1 = 3.14
x2 = 10E6

# making delta matrix
d_arr = []
for i in range(17):
  d_arr.append(10**(-1*i))
d_arr.reverse()
d_arr = np.array(d_arr)

# getting y1 and y2
y1 = np.zeros(len(d_arr))
y2 = np.zeros(len(d_arr))
for i, d in enumerate(d_arr):
  y1[i] = abs(y_o(x1, d) - y_r(x1, d))
  y2[i] = abs(y_o(x2, d) - y_r(x2, d))

# plot the results
plt.plot(d_arr, y1, label="x = " + str(x1))
plt.plot(d_arr, y2, label="x = " + str(x2))
plt.title("error in f(x,d) for varying values of x and d")
plt.xlabel("d")
plt.ylabel("f(x, d)")
plt.legend()
plt.show()
```
