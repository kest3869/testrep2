c) I we define an algorithm that checks values of n unti it finds a correct number

d) we display the absolute difference between the two nubmers to confirm they are the same within 16 digits of precision

```python
# libraries
import math

# uses a taylor series to approximate the value of e^x - 1
def taylor_approx(n, x, a):

  # first term of taylor approximation
  y_approx = math.e**x - 1

  # remaining n terms
  for i in range(1, n + 1):
    y_approx += ((math.e**a)/(math.factorial(n)))*(x - a)**n

  return y_approx

# parameters
tol = 1E-16
y_true = 9.999999995000000E-10
x = 9.999999995000000E-10

# looks at smaller and smaller n until it finds an approximation within tol
n = 1
y_approx = taylor_approx(n, x, 1)
while abs(y_true - y_approx) > tol:
  n+=1
  y_approx = taylor_approx(n, x, 1)

# display n
print("Number of terms:", n)

# confirm they are the same for the first 16 digits
in_range = abs(y_true - y_approx) < tol
print("Approxmiation within range?", in_range)
```

```python
import numpy as np

# making y and t
t = np.arange(0, np.pi, np.pi/30)
y = np.cos(t)

# function for calculating S
def S(N):
  S = 0.0
  for i in range(N):
    S += t[i]*y[i]

  # printing the value of S
  print("the sum is:", S)

  return S
```

```python
# testing it out
S(3)
S(5)
S(10)
print()
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt

# function of x
def x(t, R, dr, f, p):
  return R*(1 + dr*np.sin(f*t + p))*np.cos(t)

# function of y
def y(t, R, dr, f, p):
  return R*(1 + dr*np.sin(f*t + p))*np.sin(t)

# parameters
num_points = 100
t_vals = np.linspace(0, 2*np.pi, num_points)
R = 1.2
dr = 0.1
f = 15
p = 0

# calculating the values of x and y
x_out = np.zeros(num_points)
y_out = np.zeros(num_points)
for i in range(num_points):
  x_out[i] = x(t_vals[i], R, dr, f, p)
  y_out[i] = y(t_vals[i], R, dr, f, p)

# plotting the outputs of x and y
plt.plot(t_vals, x_out, label='x(t)')
plt.plot(t_vals, y_out, label='y(t)')
plt.title('Plot of x(t) and y(t)')
plt.xlabel('t')
plt.ylabel('output of x(t) and y(t)')
plt.legend()
plt.show()
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import random

# function of x
def x(t, R, dr, f, p):
  return R*(1 + dr*np.sin(f*t + p))*np.cos(t)

# function of y
def y(t, R, dr, f, p):
  return R*(1 + dr*np.sin(f*t + p))*np.sin(t)

# parameters
num_points = 100
t_vals = np.linspace(0, 2*np.pi, num_points)

# for loop to make 10 plots according to specifications
for j in range(10):
  R = j
  dr = 0.05
  f = 2 + j
  p = random.uniform(0,2)

  # calculating the values of x and y
  x_out = np.zeros(num_points)
  y_out = np.zeros(num_points)
  for i in range(num_points):
    x_out[i] = x(t_vals[i], R, dr, f, p)
    y_out[i] = y(t_vals[i], R, dr, f, p)

  # adding plots to plot
  plt.plot(t_vals, x_out, label=f'x{j}(t)')
  plt.plot(t_vals, y_out, label=f'y{j}(t)')

# display final plot
plt.title('Plot of x(t) and y(t)')
plt.xlabel('t')
plt.ylabel('output of x(t) and y(t)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
